## Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## Load and Prepare Data

```python
df = pd.read_csv('googleplaystore.csv')
df.head()
```

## EDA

```python
print('Data Frame shape is:', df.shape, '\n')
print(df.info())
```

```python
df['Category'].value_counts() #number of each applications on categories 
```

```python
df['Type'].value_counts() #number of Free and Paid applications
```

```python
df['Genres'].value_counts() #number of each applications on Geners 
```

```python
pd.crosstab(df.Category , df.Type)
```

```python
pd.crosstab(df['Rating'] , df['Content Rating'])
```

```python
better_rating = df[df.Rating>=4.5] #better applications base on rating
better_rating[0:3]
```

```python
pd.crosstab(better_rating['Genres'] , better_rating['Type'])
```

```python
df[df.App.str.contains('Photo')] #applications about photos
```

```python
df.groupby('Category')['Rating'].mean().reset_index().sort_values('Rating').reset_index() #sort rates base on category
```

```python
df.query('Rating>=4.5 and Type=="Free" and Category=="BEAUTY"')[0:3]
```

## Strorytelling - Visualization

```python
sns.set_context('notebook', font_scale=1.5 , rc={'lines.linwidth': 2.5})
```

```python
sns.displot(df, x = 'Rating', discrete=True) #as the plot shows, most rates are about 4
```

```python
sns.displot(df, x = 'Rating', kind='kde') #as the plot shows, most rates are about 4
```

```python
sns.set_theme(context='notebook', palette='Paired', font='sans-serif', font_scale=1, color_codes=True, rc=None)
sns.color_palette('Paired') #choose a palette
```

```python
sns.catplot(x='Type', hue='Category', data=df, kind='count', height=5, aspect=2) 
#as the plot shows Food and Health category is more than another categories
```

```python
#another view of up plot
g = sns.FacetGrid(data = df, col='Type')
g.map_dataframe(sns.histplot, x = 'Category', binwidth=2)
g.set_axis_labels('Category', 'COUNT')
g.add_legend()
```
